```python
Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
they may impact your analysis.
```

```python
#CHecking for outliers by using IQR method

import pandas as pd

# Load the flight price dataset into a pandas dataframe
df = pd.read_excel('flight_price (1).xlsx')

# Calculate the 25th and 75th percentiles and the IQR
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3 - q1

# Identify outliers outside the range of 1.5 times the IQR
outliers = df[(df['Price'] < q1 - 1.5 * iqr) | (df['Price'] > q3 + 1.5 * iqr)]
print(outliers)
```

```python
Conslusion:- Yes , there are many outliers in the dataset.
```

```python
Effect Of Oultiers in the analysis:-

Outliers can affect the analysis of a dataset in several ways. One way is that they can skew the distribution of the data,
making it difficult to accurately describe the center and spread of the data. This can affect the calculation of summary
statistics such as the mean, which may be biased by the presence of outliers.

Outliers can also affect the results of statistical tests, as many tests assume that the data is normally distributed and
free from outliers. If outliers are present in the data, this assumption may not hold, and the results of the test may be
invalid or misleading.

In addition, outliers can also affect the interpretation of patterns and relationships in the data. If outliers are 
included in the analysis, they may distort the overall patterns and trends in the data, making it difficult to identify
meaningful relationships between variables.

Therefore, it is important to carefully consider the presence of outliers in a dataset and their potential impact on the
analysis. Depending on the nature of the data and the research question, it may be appropriate to remove outliers from 
the dataset or to use statistical methods that are robust to the presence of outliers.
```

```python
#Explaanation :- I am creating four new columns 1.Date 2.Month 3.Seasons 4.	Mean_Price_According_Season
#After creating these columns , i used two columns 'Seasons' and 'Mean_Price_According_Season' for further analysis 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the Flight Price dataset into a pandas dataframe
df = pd.read_excel('flight_price (1).xlsx')

# Convert the "Date" column to a datetime object for month value 
df['Date'] = pd.to_datetime(df['Date_of_Journey'])
# Extract the month from the "Date" column
df['Month'] = df['Date'].dt.month

# Create a dictionary that maps months to seasons : (Feb-March)-Spring Season , (April,May,June)-Summer Season 
#(July,August,September)-Monsoon Season ,(October,November,December,January)-Winter Season

seasons_dict = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 
                6: 'Summer', 7: 'Monsoon', 8: 'Monsoon', 9: 'Monsoon', 10: 'Winter',
                11: 'Winter', 12: 'Winter'}

# Map the months to seasons using the seasons_dict
df['Seasons'] = df['Month'].map(seasons_dict)

# Filter the dataframe to only include flights in particular season
winter_df = df[df['Month'].isin([10, 11, 12, 1])]
spring_df = df[df['Month'].isin([2, 3])]
summer_df = df[df['Month'].isin([4,5,6])]
monsoon_df = df[df['Month'].isin([7,8,9])]

# Calculate the mean price for the All seasons
Mean_Price_in_Winter_Season = winter_df['Price'].mean()
Mean_Price_in_Spring_Season = spring_df['Price'].mean()
Mean_Price_in_Summer_Season = summer_df['Price'].mean()
Mean_Price_in_Monsoon_Season = monsoon_df['Price'].mean()

# Create a dictionary that maps seasons to mean prices
season_mean_prices = {'Winter':Mean_Price_in_Winter_Season , 'Spring':Mean_Price_in_Spring_Season, 'Summer': Mean_Price_in_Summer_Season, 'Monsoon': Mean_Price_in_Monsoon_Season}

# Map the seasons to mean prices using the season_mean_prices dictionary
df['Mean_Price_According_Season'] = df['Seasons'].map(season_mean_prices)

plt.figure(figsize=(12,6))
sns.boxplot(x='Seasons',y='Mean_Price_According_Season',data=df)
```

```python
Observation:- From the above boxplot graph , i show to boss a  conclusion that in Winter Season , the flight prices are 
high and in Spring Season the flight prices are lowest .
So to boost sales of travel packeges , travel agency should focus on promoting Winter travel packages to customers.
and in Spring Season Travel agency should offer promotions or discounts for travel during those times.
```

```python
Observation:- The following dataset has 1474 missing values in Rating column,1 missing value in Type column , 1 missing value 
in Content rating , 8 missing values in Current Ver and 3 missing values in Android Version

Missing values can have a significant impact on data analysis, as they can lead to biased or inaccurate results if not handled
properly. Here are some ways in which missing values can impact your analysis:

1.Reduced Sample Size: If a large number of observations have missing values for one or more variables, it can significantly 
reduce the sample size available for analysis. This can affect the statistical power of your analysis, and reduce the accuracy 
of your results.

2.Biased Results: If the missing values are not randomly distributed across the dataset, they can lead to biased results. For 
example, if the missing values are more common in a certain subset of the data, such as in a particular category or time
period, the analysis may be skewed towards that subset of the data.

3.Incorrect Analysis: If the missing values are not handled properly, they can lead to incorrect or misleading analysis. For\
example, if missing values are simply dropped or imputed with mean values, this can affect the distribution of the data and 
the validity of statistical tests.

To mitigate the impact of missing values on analysis, there are several strategies that can be used:

1.Imputation: Missing values can be imputed using statistical methods such as mean, median, or regression imputation. This can
help preserve the sample size and reduce bias in the results.

2.Subsetting: If missing values are concentrated in a specific subset of the data, such as a certain time period or category, 
the analysis can be restricted to the subset of the data that has complete information.

3.Dropping: If missing values are rare and randomly distributed, it may be appropriate to simply drop the observations with 
missing values. However, this should be done with caution, as it can reduce the sample size and affect the statistical power 
of the analysis.

In summary, missing values can have a significant impact on data analysis, and it is important to handle them properly to 
ensure the accuracy and validity of the results. Different strategies can be used depending on the nature and extent of the 
missing values in the dataset.
```

```python
Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship
```

```python
#importing required libraries..
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#loading dataset
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

#cleaning the required columns
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].str.replace('+','')
df['Size']=df['Size'].str.replace(',','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
df['Size']=df['Size'].astype(float)
df['Rating']=df['Rating'].astype(float)

#plotting on scatter plot
plt.scatter(df['Size'],df['Rating'])
plt.xlabel('Size of App in KB')
plt.ylabel('App Rating')
plt.title('Relationship between Size and Rating between App')
plt.xticks(rotation=90)
plt.show()
```

```python
Observation:-The scatter plot suggests that there is no strong relationship between the size of an app and its rating.
There are apps of all sizes that have high ratings, and apps of all sizes that have low ratings. However, there appears to be
a slightly higher concentration of apps with high ratings in the smaller size ranges, and a slightly higher concentration of
apps with low ratings in the larger size ranges.

It's important to note that the scatter plot only shows a visual representation of the relationship between these two variables
and doesn't indicate any causal relationship between them. Further analysis, such as correlation analysis, may be necessary to
determine the strength of the relationship between the size of an app and its rating.
```
