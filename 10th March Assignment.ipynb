```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Ans:-Estimation statistics is a branch of statistics that deals with estimating unknown population parameters based on sample
data. In inferential statistics, which is a part of estimation statistics, we use sample data to make inferences about a 
population.

- Point estimate refers to a single value that is used to estimate an unknown population parameter. It is the best estimate of
the parameter based on the available sample data. For example, if we are interested in estimating the mean height of all adults
in a certain city, we can take a sample of heights from a random sample of individuals and calculate the sample mean. This sample
mean would be the point estimate of the population mean.

- Interval estimate, on the other hand, provides a range of values within which the population parameter is likely to fall, 
along with a level of confidence. It accounts for the uncertainty associated with estimating a population parameter from a 
sample. The most common type of interval estimate is the confidence interval, which provides a range of values within which 
the population parameter is estimated to fall with a specified level of confidence. For example, a 95% confidence interval for
the population mean height would provide a range of values within which the true population mean height is estimated to fall
with 95% confidence.

In summary, point estimate is a single value used to estimate a population parameter, while interval estimate provides a range 
of values along with a level of confidence within which the population parameter is likely to fall.
```

```python
Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.
```

```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    """
    Estimates the population mean using a sample mean, sample standard deviation, and sample size.
    
    Args:
        sample_mean (float): Sample mean
        sample_std (float): Sample standard deviation
        sample_size (int): Sample size
    
    Returns:
        float: Estimated population mean
    """
    # Calculate the standard error (standard deviation of the sample mean)
    standard_error = sample_std / math.sqrt(sample_size)
    
    # Calculate the margin of error (critical value times standard error)
    # Assuming a 95% confidence interval (i.e., alpha = 0.05) for simplicity
    critical_value = 1.96  # Z-score for a 95% confidence interval
    margin_of_error = critical_value * standard_error
    
    # Calculate the estimated population mean by adding the margin of error to the sample mean
    estimated_population_mean = sample_mean + margin_of_error
    
    return estimated_population_mean
```

```python
Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
```

```python
import numpy as np
from scipy.stats import t

def conduct_two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Conducts a two-sample t-test on the difference between two population means.
    
    Args:
        sample1 (array-like): Sample from population 1
        sample2 (array-like): Sample from population 2
        alpha (float, optional): Significance level for the hypothesis test. Defaults to 0.05.
    
    Returns:
        tuple: t-statistic, degrees of freedom, p-value, and whether to reject the null hypothesis
    """
    # Calculate sample statistics
    n1 = len(sample1)
    n2 = len(sample2)
    x1_bar = np.mean(sample1)
    x2_bar = np.mean(sample2)
    s1 = np.std(sample1, ddof=1)
    s2 = np.std(sample2, ddof=1)
    
    # Calculate pooled standard deviation and standard error
    pooled_std = np.sqrt(((n1 - 1) * s1 ** 2 + (n2 - 1) * s2 ** 2) / (n1 + n2 - 2))
    standard_error = pooled_std * np.sqrt(1 / n1 + 1 / n2)
    
    # Calculate t-statistic
    t_statistic = (x1_bar - x2_bar) / standard_error
    
    # Calculate degrees of freedom
    df = n1 + n2 - 2
    
    # Calculate p-value
    p_value = 2 * (1 - t.cdf(np.abs(t_statistic), df=df))
    
    # Determine whether to reject the null hypothesis
    reject_null = p_value < alpha
    
    return t_statistic, df, p_value, reject_null

# Example usage
# Generate example sample data for two populations
np.random.seed(0)
population1 = np.random.normal(loc=10, scale=2, size=100)
population2 = np.random.normal(loc=12, scale=3, size=120)

# Generate example samples from the populations
sample1 = np.random.choice(population1, size=30, replace=False)
sample2 = np.random.choice(population2, size=40, replace=False)

# Conduct the two-sample t-test
t_statistic, df, p_value, reject_null = conduct_two_sample_t_test(sample1, sample2)

# Print results
print("Sample 1 Mean: {:.2f}".format(np.mean(sample1)))
print("Sample 2 Mean: {:.2f}".format(np.mean(sample2)))
print("t-statistic: {:.2f}".format(t_statistic))
print("Degrees of Freedom: {}".format(df))
print("p-value: {:.4f}".format(p_value))
print("Reject Null Hypothesis: {}".format(reject_null))
```

```python
Q6: What is a null and alternative hypothesis? Give some examples.
Ans:-In statistics, a null hypothesis (H0) is a statement that suggests that there is no significant relationship or difference
between variables or populations. It assumes that any observed differences or relationships are due to chance or random 
variation. On the other hand, an alternative hypothesis (Ha or H1) is a statement that contradicts or opposes the null 
hypothesis, suggesting that there is a significant relationship or difference between variables or populations.

Here are some examples of null and alternative hypotheses:

1.Example for comparing means:

Null hypothesis (H0): The average test scores of students who study with music and those who study without music are the same.
Alternative hypothesis (Ha): The average test scores of students who study with music are different from those who study 
without music.

2.Example for comparing proportions:
Null hypothesis (H0): The proportion of people who prefer brand A is equal to the proportion of people who prefer brand B.
Alternative hypothesis (Ha): The proportion of people who prefer brand A is different from the proportion of people who prefer
brand B.

3.Example for testing correlation:
Null hypothesis (H0): There is no significant correlation between age and income.
Alternative hypothesis (Ha): There is a significant correlation between age and income.

4.Example for testing the effectiveness of a new drug:
Null hypothesis (H0): The new drug has no effect on reducing blood pressure.
Alternative hypothesis (Ha): The new drug is effective in reducing blood pressure.

5.Example for testing the impact of a marketing campaign:
Null hypothesis (H0): The marketing campaign has no significant impact on sales.
Alternative hypothesis (Ha): The marketing campaign has a significant impact on sales.

In hypothesis testing, the null hypothesis is usually assumed to be true unless there is sufficient evidence to reject it in 
favor of the alternative hypothesis based on statistical analysis of data. The choice of null and alternative hypotheses 
depends on the research question or problem being investigated, and they are used to make conclusions about the relationships 
or differences between variables or populations based on available data and statistical analysis.
```

```python
Q7: Write down the steps involved in hypothesis testing.
Ans:-Hypothesis testing involves several steps to draw conclusions about the relationships or differences between variables or
populations based on available data. The general steps in hypothesis testing are as follows:

1.State the null and alternative hypotheses: Clearly define the null hypothesis (H0) and the alternative hypothesis (Ha or H1)
based on the research question or problem being investigated. The null hypothesis assumes that there is no significant 
relationship or difference between variables or populations, while the alternative hypothesis suggests that there is a 
significant relationship or difference.

2.Choose the appropriate test statistic: Select the appropriate statistical test based on the type of data and the research
question. Commonly used tests include t-test, z-test, chi-squared test, ANOVA, etc.

3.Set the significance level: Choose the significance level (also known as alpha level) that determines the threshold for 
rejecting the null hypothesis. The significance level is typically set at 0.05, which means that there is a 5% chance of
making a Type I error (rejecting a true null hypothesis).

4.Collect and analyze the data: Collect the relevant data and perform the appropriate statistical analysis using the chosen 
test statistic. Calculate the test statistic value from the sample data.

5.Calculate the p-value: Based on the test statistic value, calculate the p-value, which represents the probability of 
obtaining a test statistic value as extreme or more extreme than the observed value, assuming the null hypothesis is true. The
p-value is used to determine the strength of evidence against the null hypothesis.

6.Make a decision: Compare the p-value with the significance level. If the p-value is less than the significance level, reject
the null hypothesis in favor of the alternative hypothesis, indicating that there is sufficient evidence to support the 
alternative hypothesis. If the p-value is greater than the significance level, fail to reject the null hypothesis, indicating 
that there is not enough evidence to support the alternative hypothesis.

7.Interpret the results: Based on the decision made, interpret the results in the context of the research question or problem 
being investigated. Draw conclusions and make inferences about the relationships or differences between variables or 
populations based on the statistical analysis.

8.Consider limitations: Consider the limitations of the hypothesis testing process, such as sample size, assumptions of the 
statistical test, potential confounding variables, and generalizability of the results to the broader population. Interpret 
the results with caution and acknowledge any limitations or uncertainties in the conclusions drawn.

9.Report the findings: Finally, report the findings of the hypothesis test, including the decision made, the interpretation of
results, and any limitations or recommendations for future research, in a clear and concise manner.

It's important to note that hypothesis testing is a statistical tool that provides evidence, but it does not prove or disprove 
a hypothesis with certainty. It helps researchers make conclusions based on the available data and statistical analysis, and it
should be interpreted and applied in the context of the specific research question or problem being investigated.
```

```python
Q8. Define p-value and explain its significance in hypothesis testing.
Ans:-The p-value, also known as the probability value, is a measure of the strength of evidence against the null hypothesis in 
hypothesis testing. It represents the probability of obtaining a test statistic value as extreme or more extreme than the
observed value, assuming the null hypothesis is true. In other words, it quantifies the likelihood of observing the observed
data, or data more extreme, under the assumption that the null hypothesis is correct.

The significance of the p-value in hypothesis testing is as follows:

1.Decision making: The p-value is used as a criterion to make a decision about whether to reject or fail to reject the null 
hypothesis. If the p-value is less than the significance level (often set at 0.05), it is considered statistically significant,
and the null hypothesis is rejected in favor of the alternative hypothesis, indicating that there is sufficient evidence to
support the alternative hypothesis. If the p-value is greater than the significance level, the null hypothesis is not rejected,
and it is concluded that there is not enough evidence to support the alternative hypothesis.

2.Interpretation of results: The p-value helps in interpreting the results of hypothesis testing. A smaller p-value indicates
stronger evidence against the null hypothesis, suggesting that the observed data are unlikely to occur by chance alone. On the 
other hand, a larger p-value indicates weaker evidence against the null hypothesis, suggesting that the observed data could be
reasonably explained by random chance.

3.Confidence in conclusions: The p-value provides a measure of the strength of evidence against the null hypothesis, which 
helps in assessing the confidence level of the conclusions drawn from the hypothesis test. A smaller p-value indicates higher 
confidence in rejecting the null hypothesis and supporting the alternative hypothesis, while a larger p-value indicates lower
confidence in rejecting the null hypothesis.

4.Reproducibility of results: The p-value allows for the replication of the hypothesis test by other researchers, as it 
provides a quantitative measure of the likelihood of obtaining similar results in future studies. If the p-value is small, it
suggests that the findings are less likely to be due to chance and can be replicated, increasing the reliability and
reproducibility of the results.

5.Limitations and interpretation: It's important to interpret the p-value in the context of the research question, study 
design, and assumptions of the statistical test used. The p-value does not provide information about the practical significance
or effect size of the observed relationship or difference. Additionally, the p-value is influenced by sample size, and a larger
sample size may result in a smaller p-value even for a small effect size. Therefore, careful interpretation of the p-value, 
along with other relevant statistical measures and contextual factors, is crucial in drawing meaningful conclusions from
hypothesis testing.

In summary, the p-value is a critical statistical measure in hypothesis testing that helps in making decisions, interpreting 
results, assessing confidence, ensuring reproducibility, and understanding the limitations of the findings.
```

```python
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
```

```python
import scipy.stats as stats

# Generate two random samples
sample1 = stats.norm.rvs(loc=10, scale=2, size=100)
sample2 = stats.norm.rvs(loc=12, scale=2, size=100)

# Calculate the t-statistic and p-value for the two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Set the significance level
alpha = 0.05

# Print the results
print("Two-Sample T-Test Results:")
print("Sample 1 Mean:", sample1.mean())
print("Sample 2 Mean:", sample2.mean())
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
if p_value < alpha:
    print("Reject the null hypothesis (population means are equal)")
else:
    print("Fail to reject the null hypothesis (population means are equal)")
```

```python
Q11: What is Student’s t distribution? When to use the t-Distribution.
Ans:-Student's t-distribution, also known as simply the t-distribution, is a probability distribution that is used in statistical
inference when estimating the population mean of a small sample from a normally distributed population, or when comparing the 
means of two small samples.

The t-distribution is similar to the standard normal distribution (Z-distribution), but it has thicker tails, which makes it 
more suitable for small sample sizes where the assumption of normality may not hold or the population standard deviation is 
unknown. The t-distribution is characterized by its degrees of freedom (df), which determines the shape of the distribution.

The t-distribution is commonly used in the following situations:

1.Small sample sizes: When the sample size is small (typically less than 30), the t-distribution is used instead of the standard
normal distribution for estimating the population mean or conducting hypothesis tests. This is because the standard normal distribution
assumes a large sample size and may not be accurate for small samples due to the thicker tails of the t-distribution.

2.Unknown population standard deviation: When the population standard deviation is unknown, the t-distribution is used for 
estimating the population mean or conducting hypothesis tests. This is because the t-distribution accounts for the uncertainty
introduced by estimating the standard deviation from the sample data.

3.Comparing two small samples: When comparing the means of two small samples (typically less than 30) using a hypothesis test, 
the t-distribution is used instead of the standard normal distribution. This is known as the two-sample t-test for independent 
samples or paired samples, depending on the study design.

In summary, the t-distribution is used in statistical inference when working with small sample sizes or when the population
standard deviation is unknown, and it is commonly used for estimating population means or comparing means of two small samples.
```

```python
Q12: What is t-statistic? State the formula for t-statistic.
Ans:-The t-statistic is a measure that quantifies the difference between the sample mean and a hypothesized population mean in 
terms of the standard error of the sample mean. It is used in hypothesis testing to assess whether the difference between the
sample mean and the hypothesized population mean is statistically significant.

The formula for calculating the t-statistic is as follows:

t = (x̄ - μ) / (s / √n)

Where:

x̄ is the sample mean
μ is the hypothesized population mean
s is the sample standard deviation
n is the sample size
In this formula, the numerator (x̄ - μ) represents the difference between the sample mean and the hypothesized population mean. 
The denominator (s / √n) represents the standard error of the sample mean, where s is the sample standard deviation and √n is 
the square root of the sample size.

The t-statistic measures how many standard errors the sample mean is away from the hypothesized population mean. A larger absolute
value of the t-statistic indicates a larger difference between the sample mean and the hypothesized population mean, which may
provide evidence against the null hypothesis in a hypothesis test. The t-statistic is used in conjunction with the t-distribution
to calculate the p-value, which is used to make a decision on whether to reject or fail to reject the null hypothesis in 
hypothesis testing.
```
