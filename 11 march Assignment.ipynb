```python
Q2: Differentiate between one-tailed and two-tailed tests.
Ans:-In statistics, one-tailed and two-tailed tests refer to the directionality of a statistical hypothesis and the corresponding
critical region in a hypothesis test. 

Here's how they differ:

One-tailed test:

Also known as a directional test.
The null hypothesis (H0) specifies the direction of the effect, and the alternative hypothesis (Ha) states the opposite direction.
The critical region is located on one side of the distribution, either in the upper tail or the lower tail.
It tests for the possibility of an effect in only one direction.
Used when there is a specific prediction or expectation about the direction of the effect.
Example: Testing whether a new drug increases average response time, where the research hypothesis is that the drug increases
response time, and the null hypothesis is that it does not.

Two-tailed test:

Also known as a non-directional test.
The null hypothesis (H0) states that there is no effect or difference, and the alternative hypothesis (Ha) states that there
is an effect or difference, but does not specify the direction.
The critical region is located on both sides of the distribution, in both the upper and lower tails.
It tests for the possibility of an effect in either direction.
Used when there is no specific prediction or expectation about the direction of the effect, or when researchers are interested 
in detecting any kind of difference.
Example: Testing whether a new teaching method affects exam scores, where the research hypothesis is that the teaching method
affects scores, but does not specify whether scores will increase or decrease.

In summary, one-tailed tests are used when there is a specific prediction about the direction of the effect, while two-tailed
tests are used when there is no specific prediction or when researchers are interested in detecting any kind of difference.
One-tailed tests have a single critical region, while two-tailed tests have two critical regions on both sides of the distribution.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
Ans:-Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing:

Type 1 error (False Positive):
Also known as an alpha error or a level of significance error.
Occurs when a null hypothesis (H0) is rejected, even though it is true.
In other words, it is a false claim of a significant effect or difference.
The probability of making a Type 1 error is denoted by the significance level (α), which is predetermined by the researcher.
A lower significance level (e.g., α = 0.01) reduces the chances of making a Type 1 error, but increases the risk of Type 2 
error.

Example scenario: A pharmaceutical company conducts a clinical trial to test the effectiveness of a new drug in reducing 
cholesterol levels. The null hypothesis is that the drug has no effect, but due to a Type 1 error, the researchers incorrectly 
conclude that the drug is effective when it is not, leading to a false claim of its effectiveness.

Type 2 error (False Negative):
Also known as a beta error or a power error.
Occurs when a null hypothesis (H0) is failed to be rejected, even though it is false.
In other words, it is a failure to detect a true effect or difference.
The probability of making a Type 2 error is denoted by the symbol β (beta), and it is inversely related to the power of a 
statistical test, which is the probability of correctly rejecting a false null hypothesis.
A higher power (e.g., 1-β = 0.8) reduces the chances of making a Type 2 error, but increases the risk of Type 1 error.

Example scenario: A researcher conducts an experiment to test whether a new teaching method improves students' test scores.
The null hypothesis is that the teaching method has no effect, but due to a Type 2 error, the researcher fails to detect a true
effect and erroneously concludes that the teaching method is not effective.

In summary, Type 1 error is a false positive, where a true null hypothesis is rejected, while Type 2 error is a false negative,
where a false null hypothesis is failed to be rejected. Researchers need to carefully consider the significance level and power
of their statistical tests to minimize the risk of both types of errors in hypothesis testing.
```

```python
Q4: Explain Bayes's theorem with an example.
Ans:-Bayes's theorem is a mathematical formula that allows for the revision of probabilities based on new information. It is 
commonly used in statistics and probability theory to update the probability of an event occurring, given new evidence or
information. Bayes's theorem is expressed as:

P(A|B) = P(B|A) * P(A) / P(B)

where:

P(A|B) is the conditional probability of event A occurring given that event B has occurred.
P(B|A) is the conditional probability of event B occurring given that event A has occurred.
P(A) and P(B) are the probabilities of events A and B occurring, respectively.
Here's an example to illustrate Bayes's theorem:

Suppose you are a doctor trying to diagnose a rare disease that affects 1 in 1000 people. You have a test that is 99% accurate,
meaning that it correctly identifies the disease in 99% of cases and gives a false positive in 1% of cases. You have a patient
who tests positive for the disease with the test. What is the probability that the patient actually has the disease?

Let's use Bayes's theorem to calculate it:

Let event A be the patient having the disease (1 in 1000 people).
Let event B be the patient testing positive for the disease with the test.

Given:
P(A) = 1/1000 (probability of the patient having the disease)
P(B|A) = 0.99 (probability of the patient testing positive given that they have the disease)
P(B|¬A) = 0.01 (probability of the patient testing positive given that they do not have the disease)

We need to find P(A|B), the probability of the patient actually having the disease given that they tested positive with the test.

Using Bayes's theorem:
P(A|B) = P(B|A) * P(A) / P(B|A) * P(A) + P(B|¬A) * P(¬A)
P(A|B) = 0.99 * (1/1000) / (0.99 * (1/1000) + 0.01 * (999/1000))

After calculating the above expression, we find that P(A|B) is approximately 0.09 or 9%. So, even though the patient tested 
positive for the disease, the probability of them actually having the disease is only 9% due to the rarity of the disease and 
the imperfect accuracy of the test.

This example illustrates how Bayes's theorem can be used to update probabilities based on new information, such as test results,
and is commonly used in fields like medicine, epidemiology, and decision-making under uncertainty.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Ans:-A confidence interval is a statistical range that is used to estimate the unknown true value of a population parameter,
such as a mean or a proportion, based on a sample from that population. It provides a range of plausible values for the parameter
with a specified level of confidence, which represents the probability that the true parameter value falls within the interval.

Confidence intervals are often used in inferential statistics to estimate population parameters when it is not feasible or 
practical to measure the entire population. They are commonly reported with a confidence level, which is expressed as a percentage
(e.g., 95% confidence interval) and represents the probability that the true parameter value falls within the interval.

The formula for calculating a confidence interval depends on the type of parameter being estimated and the underlying distribution 
of the data. Here are the general steps to calculate a confidence interval for a population mean:

Collect a sample from the population of interest.
Calculate the sample statistics, such as the sample mean (x̄) and the sample standard deviation (s).
Determine the desired confidence level (e.g., 95% confidence interval) and find the corresponding critical value from a standard 
normal distribution or a t-distribution, depending on the sample size and whether the population standard deviation is known.

Use the formula to calculate the confidence interval:
Confidence Interval = Sample statistic ± (Critical value * Standard error)

where the standard error is the standard deviation of the sample statistic, which is calculated as the sample standard deviation
divided by the square root of the sample size (s / √n).

Here's an example of calculating a 95% confidence interval for a population mean:

Suppose you want to estimate the average height of a certain population of adults. You collect a random sample of 100 individuals
from this population and measure their heights. The sample mean height is 65 inches, and the sample standard deviation is 
3 inches. You want to calculate a 95% confidence interval for the true population mean height.

Given:
Sample size (n) = 100
Sample mean (x̄) = 65 inches
Sample standard deviation (s) = 3 inches
Confidence level = 95%

Steps:

1.Determine the critical value for a 95% confidence interval for a sample size of 100 from a standard normal distribution or 
t-distribution (depending on assumptions about the population and sample size).

2.Calculate the standard error: s / √n = 3 / √100 = 0.3 inches.
3.Use the formula to calculate the confidence interval:
  Confidence Interval = 65 ± (Critical value * 0.3)

After determining the critical value and plugging in the values, you would get the lower and upper bounds of the confidence 
interval, which would give you the range of plausible values for the true population mean height with a 95% confidence level.

Note that as the confidence level increases, the width of the confidence interval also increases, reflecting a wider range of 
plausible values for the parameter. Similarly, as the sample size increases, the standard error decreases, leading to a 
narrower confidence interval and a more precise estimate of the parameter.
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.

Ans:-Bayes' Theorem is a mathematical formula that allows us to update the probability of an event occurring based on new information.
It is expressed as:

P(A|B) = P(B|A) * P(A) / P(B)

where:

P(A|B) is the conditional probability of event A occurring given that event B has occurred.
P(B|A) is the conditional probability of event B occurring given that event A has occurred.
P(A) is the prior probability of event A occurring.
P(B) is the prior probability of event B occurring.
Let's illustrate this with an example problem:

Example Problem:
Suppose you are a doctor and you are diagnosing a rare disease in patients. You know that the prevalence of the disease in the 
general population is 1%, and that the test for the disease has a sensitivity of 90% and a specificity of 95%. You receive a 
positive test result for a patient. What is the probability that the patient actually has the disease?

Solution:
In this problem, we are trying to find the probability that the patient actually has the disease given a positive test result,
which is P(Disease|Positive). We are given the following information:

P(Disease) = 1% = 0.01 (prior probability of the disease occurring)
P(Positive|Disease) = Sensitivity = 90% = 0.9 (probability of a positive test result given the patient has the disease)
P(Positive|No Disease) = 1 - Specificity = 1 - 95% = 5% = 0.05 (probability of a positive test result given the patient does 
                                                                not have the disease)

We can now use Bayes' Theorem to calculate the probability of the patient actually having the disease given a positive test 
result:

P(Disease|Positive) = P(Positive|Disease) * P(Disease) / P(Positive)

P(Positive) can be calculated using the law of total probability as:

P(Positive) = P(Positive|Disease) * P(Disease) + P(Positive|No Disease) * P(No Disease)

P(No Disease) is the complement of P(Disease), i.e., 1 - P(Disease).

Plugging in the given values, we get:

P(Positive) = 0.9 * 0.01 + 0.05 * (1 - 0.01) = 0.0585

Now, we can plug this value into Bayes' Theorem to calculate P(Disease|Positive):

P(Disease|Positive) = P(Positive|Disease) * P(Disease) / P(Positive)
= 0.9 * 0.01 / 0.0585
≈ 0.1538 or 15.38%

So, the probability that the patient actually has the disease given a positive test result is approximately 15.38%. This 
calculation takes into account the prior probability of the disease, as well as the sensitivity and specificity of the test,
and provides an updated probability based on the new evidence of a positive test result.
```

```python
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

Ans:-The margin of error in a confidence interval is a measure of the amount of uncertainty or imprecision associated with the 
estimated parameter (e.g., mean, proportion) based on a sample of data. It represents the range within which the true population
parameter is likely to fall with a certain level of confidence.

The margin of error is calculated as half of the width of the confidence interval. In other words, it is the distance between 
the estimated parameter (e.g., sample mean) and the upper or lower limit of the confidence interval.

Sample size has an inverse relationship with the margin of error. As the sample size increases, the margin of error decreases,
and vice versa. A larger sample size generally results in a smaller margin of error, indicating a more precise estimate of the 
true population parameter.

Here's an example to illustrate this:

Scenario: A political poll is conducted to estimate the proportion of voters who support a particular candidate. The poll is 
conducted with a sample of 500 voters, and a 95% confidence level is chosen.

Solution:
Let's assume that the estimated proportion of voters who support the candidate is 0.60 based on the sample, and the standard 
error (a measure of variability) is calculated to be 0.02.

Using the formula for the margin of error:
Margin of Error = Critical Value * Standard Error

For a 95% confidence level, the critical value for a proportion estimation is approximately 1.96 (for a normal distribution).

Margin of Error = 1.96 * 0.02
= 0.0392 (rounded to four decimal places)

So, the margin of error is approximately 0.0392 or 3.92%.

Now, let's consider a scenario where the sample size is increased to 1000 voters, keeping everything else the same.

Using the same formula:
Margin of Error = 1.96 * 0.02 / √1000
= 0.0063 (rounded to four decimal places)

So, the margin of error is approximately 0.0063 or 0.63%.

As we can see, with a larger sample size (1000 voters), the margin of error decreases from 3.92% to 0.63%, indicating a more
precise estimate of the true proportion of voters who support the candidate. This illustrates how sample size affects the margin 
of error in a confidence interval, with larger sample sizes resulting in smaller margins of error and more precise estimates.

```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

Ans:-To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence
level using a t-test, we need to set up the null hypothesis (H0) and the alternative hypothesis (H1) as follows:

Null hypothesis (H0): The weight loss drug is not significantly effective. The population mean weight loss is equal to or
less than zero.
Alternative hypothesis (H1): The weight loss drug is significantly effective. The population mean weight loss is greater
than zero.

We will use a one-sample t-test, as we are comparing the sample mean to a known value (zero) and do not have information 
about the population standard deviation.

Given information:
Sample size (n) = 50
Sample mean (x̄) = 6 pounds
Sample standard deviation (s) = 2.5 pounds
Confidence level = 95%

Let's calculate the test statistic (t-value) and the critical value using the t-distribution table or a statistical 
software, and then make a decision based on the results.

Step 1: Calculate the test statistic (t-value)
The formula for the t-value in a one-sample t-test is:
t = (x̄ - μ) / (s / sqrt(n))
where:
x̄ = sample mean
μ = hypothesized population mean (in this case, zero)
s = sample standard deviation
n = sample size

Plugging in the values:
x̄ = 6
μ = 0
s = 2.5
n = 50

t = (6 - 0) / (2.5 / sqrt(50))
t = 6 / (2.5 / 7.0711)
t = 6 / 0.3536
t ≈ 16.96 (rounded to two decimal places)

Step 2: Calculate the critical value
Since we are conducting a one-tailed test (we are interested in whether the population mean weight loss is greater than zero),
and our confidence level is 95%, we can find the critical value (t-critical) using a t-distribution table or a statistical
software.

For a 95% confidence level with 49 degrees of freedom (n-1 = 50-1 = 49), the t-critical value is approximately 1.676
(rounded to three decimal places) for a one-tailed test in the positive direction.

Step 3: Make a decision
Compare the calculated test statistic (t-value) with the critical value (t-critical):
t-value = 16.96
t-critical = 1.676

If the calculated t-value is greater than the critical value, we reject the null hypothesis (H0) and conclude that the 
weight loss drug is significantly effective. If the calculated t-value is less than or equal to the critical value, we do 
not reject the null hypothesis and do not conclude that the weight loss drug is significantly effective.

In this case, 16.96 > 1.676, so we would reject the null hypothesis and conclude that the weight loss drug is significantly 
effective at a 95% confidence level. This suggests that the weight loss drug has a statistically significant effect on
weight loss based on the given sample data.
```

```python
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean.

Ans:-To calculate the 90% confidence interval for the true population mean, we can use the formula for a confidence 
interval for a population mean when the population standard deviation is known.

Given information:
Population mean (μ) = 60
Population standard deviation (σ) = 8
Sample size (n) = 50
Sample mean (x̄) = 65
Confidence level = 90%

Step 1: Determine the confidence level
The confidence level is given as 90%.

Step 2: Find the critical value
The critical value (z-score) can be found using a z-table or a statistical calculator for the given confidence level. For
a 90% confidence level, the critical value for a standard normal distribution is approximately 1.645.

Step 3: Calculate the margin of error
The margin of error (E) is calculated by multiplying the critical value by the standard deviation divided by the square
root of the sample size.
E = (z-score) * (σ / sqrt(n))

Substituting the given values, we get:
E = 1.645 * (8 / sqrt(50))
E ≈ 2.77

Step 4: Calculate the confidence interval
The confidence interval is calculated by adding and subtracting the margin of error from the sample mean.
Confidence interval = (x̄ - E, x̄ + E)

Substituting the given sample mean and the calculated margin of error, we get:
Confidence interval = (65 - 2.77, 65 + 2.77)
Confidence interval ≈ (62.23, 67.77)

Step 5: Interpret the results
The 90% confidence interval for the true population mean is approximately (62.23, 67.77). This means that we can be 90% 
confident that the true population mean falls within this interval.
```
