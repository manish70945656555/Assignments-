```python
def print_exception_hierarchy(exceptions=None, level=0):
    """
    Print the exception hierarchy for the given exceptions
    """
    if exceptions is None:
        # If no exceptions are provided, use the built-in ones
        exceptions = BaseException.__subclasses__()

    for exception in exceptions:
        print(f"{' ' * level}{exception.__name__}")
        print_exception_hierarchy(exception.__subclasses__(), level + 1)

# Print the exception hierarchy for built-in exceptions
print_exception_hierarchy()
```

```python
Q6. List down some best practices for exception handling in python
ANS:- Some best practices to follow when handling exceptions in Python:

1)Be specific: Catch only the exceptions that you expect and can handle. Avoid catching generic exceptions like Exception,
  BaseException, or SystemExit as they may mask important errors.

2)Use the finally block: When necessary, use the finally block to ensure that resources are released, regardless of
  whether an exception occurs or not. For example, closing a file or database connection in the finally block ensures 
  that the resources are properly released.

3)Keep the try block small: Put only the minimum amount of code necessary in the try block. This makes it easier to 
  isolate the code that is causing the exception and to catch the specific exception that you expect.

4)Use with statement: Use the with statement to automatically close file handlers or database connections when you are
  done with them. The with statement ensures that the resources are properly released, even if an exception occurs.

5)Provide meaningful error messages: Include detailed error messages that help the user understand what went wrong and 
 how to fix it. Be sure to log the error messages to a file or database so that you can track and diagnose problems in 
 production.

6)Use exception chaining: When an exception occurs, use exception chaining to provide information about the original error
  that caused the exception. You can do this by passing the original exception as the cause parameter when raising a new
   exception.

7)Use built-in exceptions: Use built-in exceptions whenever possible. This makes your code more readable and easier to
  maintain, as other developers will be familiar with the standard exceptions.

8)Test your exception handling: Write unit tests that cover your exception handling code. This ensures that your code
   handles all expected exceptions and that it fails gracefully when unexpected exceptions occur.
```
