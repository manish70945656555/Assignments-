```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Ans:-To calculate the Pearson correlation coefficient, we need to first compute the covariance and standard deviation 
of both variables. Here are the steps:

1.Calculate the mean of each variable.
2.For each data point, subtract the mean of its respective variable from the value of that variable.
3.Multiply these differences for each data point and add them up. This is the covariance.
4.Calculate the standard deviation for each variable by taking the square root of the sum of the squares of the differences between each data point and the mean, divided by n-1.
5.Multiply the standard deviations for both variables together.
6.Divide the covariance by the product of the standard deviations.

Let's say we have collected data from 20 students, and you find that the mean study time is 4 hours and the mean exam 
score is 80%. The covariance between study time and exam score is 60, and the standard deviation of study time is
1.5 hours while that of exam score is 10%.

So, the Pearson correlation coefficient between study time and exam score is:

r = cov(study time, exam score) / (std(study time) * std(exam score))
r = 60 / (1.5 * 10)
r = 4

A Pearson correlation coefficient of 4 indicates a strong positive linear relationship between the two variables, meaning
that students who spend more time studying tend to get higher scores on the exam. The coefficient can range from -1 to 1,
with -1 indicating a strong negative linear relationship (as one variable increases, the other decreases) and 1 indicating
a strong positive linear relationship (as one variable increases, so does the other). A coefficient of 0 indicates no linear
relationship between the two variables.
```
