```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans:- A scatter plot is a type of plot used to display the relationship between two continuous variables. 
     Each point in the plot represents a pair of values for the two variables, and the position of the point
      on the plot represents the values of the two variables for that pair.

we can generate a scatter plot using the given data and Matplotlib:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title("Relationship between x and y")
plt.xlabel("x values")
plt.ylabel("y values")

# Show the plot
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
The subplot() function in Matplotlib is used to create multiple plots in the same figure. It is useful when
you want to compare multiple plots or when you want to create a grid of plots.

Here's how we can use the subplot() function to create four line plots using the given data:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the first line plot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title("Line 1")

# Plot the second line plot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title("Line 2")

# Plot the third line plot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title("Line 3")

# Plot the fourth line plot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title("Line 4")

# Add a common x-label and y-label for all plots
fig.text(0.5, 0.04, 'X values', ha='center')
fig.text(0.04, 0.5, 'Y values', va='center', rotation='vertical')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# Show the plot
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
Ans:- A bar plot is a type of plot in which rectangular bars are used to represent the data. The height or 
      length of each bar represents the value of the corresponding data point. A bar plot is used to compare 
      different categories or groups of data, and to visualize the distribution or frequency of data.

Here's how we can use Matplotlib to create a bar plot and a horizontal bar plot using the given data:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.title("Bar Plot of Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title("Horizontal Bar Plot of Company Profit")
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
Ans:-A box plot, also known as a box-and-whisker plot, is a type of plot that shows the distribution of a
     dataset. The box in the plot represents the middle 50% of the data, with the median represented by a
    line inside the box. The whiskers extend to the minimum and maximum values of the data, excluding outliers
     which are represented as points.

    Box plots are used to compare the distribution of different datasets and to identify outliers. They can 
    also be used to visualize the spread and skewness of the data.

Here's how we can use Matplotlib to create a box plot using the given data:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
data = [box1, box2]
plt.boxplot(data)
plt.title("Box Plot of Two Datasets")
plt.xlabel("Dataset")
plt.ylabel("Values")
plt.show()
```
