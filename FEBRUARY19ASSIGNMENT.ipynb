```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
Ans:-Flask is a lightweight web framework for Python that allows you to build web applications quickly and easily.
Flask provides developers with a simple and flexible approach to building web applications and APIs by providing basic 
building blocks such as routing, request handling, and response generation.

Some advantages of using Flask include:

1.Lightweight and easy to learn: Flask is a lightweight framework that doesn't come with a lot of built-in functionality. 
  This makes it easy to learn and use, especially for beginners.

2.Flexibility: Flask is highly flexible and allows developers to use any third-party libraries or tools they want to add
  additional functionality to their application.

3.Extensibility: Flask is highly extensible, which means you can easily add new features or functionality to your
  application without having to start from scratch.

4.Customizable: Flask allows you to customize every aspect of your application, from the templates to the data storage
  layer.

5.Modular: Flask is a modular framework, which means that you can pick and choose which components you want to use and
  discard the ones you don't need. This makes it highly scalable and easy to maintain.

6.Good documentation: Flask has a very good documentation, making it easier for developers to get started with the 
 framework and find solutions to common problems.

7.Large community: Flask has a large and active community of developers who contribute to the framework, create plugins,
   and provide support to other developers. This means that you can find solutions to your problems quickly and easily
    
```

```python
Q3. What is App routing in Flask? Why do we use app routes?
Ans:-App routing in Flask refers to the process of mapping a URL or web address to a specific function or view in your 
     Flask application. In other words, app routing allows you to define which function should be called when a
     particular URL is requested by the user.

In Flask, you can define app routes using the @app.route() decorator. For example, the following code defines a route for
the URL /hello:

@app.route('/hello')
def hello():
    return 'Hello World!'

When a user requests the URL /hello, Flask will call the hello() function and return the string "Hello World!" as the 
response.

We use app routes in Flask to define the endpoints of our web application. App routes allow users to navigate through 
our application by requesting different URLs. By defining app routes, we can create a RESTful API or a web application 
that can perform different functions based on the URL requested by the user.

For example, we can define app routes for different pages of a web application, such as a login page, a dashboard page,
or a settings page. Each route will correspond to a specific function that will handle the logic for that page.

In summary, app routing is an essential part of Flask that allows us to define the endpoints of our web application and 
map them to specific functions. This makes it easy to create complex web applications and RESTful APIs using Flask.
```
